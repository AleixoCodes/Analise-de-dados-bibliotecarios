## **Importando Dados**

```python
import pandas as pd
```

```python
emprestimos2010_1 = pd.read_csv('/content/drive/MyDrive/emprestimos/emprestimos-20101.csv')
emprestimos2010_2 = pd.read_csv('/content/drive/MyDrive/emprestimos/emprestimos-20102.csv')
emprestimos2011_1 = pd.read_csv('/content/drive/MyDrive/emprestimos/emprestimos-20111.csv')
emprestimos2011_2 = pd.read_csv('/content/drive/MyDrive/emprestimos/emprestimos-20112.csv')
emprestimos2012_1 = pd.read_csv('/content/drive/MyDrive/emprestimos/emprestimos-20121.csv')
emprestimos2012_2 = pd.read_csv('/content/drive/MyDrive/emprestimos/emprestimos-20122.csv')
emprestimos2013_1 = pd.read_csv('/content/drive/MyDrive/emprestimos/emprestimos-20131.csv')
emprestimos2013_2 = pd.read_csv('/content/drive/MyDrive/emprestimos/emprestimos-20132.csv')
emprestimos2014_1 = pd.read_csv('/content/drive/MyDrive/emprestimos/emprestimos-20141.csv')
emprestimos2014_2 = pd.read_csv('/content/drive/MyDrive/emprestimos/emprestimos-20142.csv')
emprestimos2015_1 = pd.read_csv('/content/drive/MyDrive/emprestimos/emprestimos-20151.csv')
emprestimos2015_2 = pd.read_csv('/content/drive/MyDrive/emprestimos/emprestimos-20152.csv')
emprestimos2016_1 = pd.read_csv('/content/drive/MyDrive/emprestimos/emprestimos-20161.csv')
emprestimos2016_2 = pd.read_csv('/content/drive/MyDrive/emprestimos/emprestimos-20162.csv')
emprestimos2017_1 = pd.read_csv('/content/drive/MyDrive/emprestimos/emprestimos-20171.csv')
emprestimos2017_2 = pd.read_csv('/content/drive/MyDrive/emprestimos/emprestimos-20172.csv')
emprestimos2018_1 = pd.read_csv('/content/drive/MyDrive/emprestimos/emprestimos-20181.csv')
emprestimos2018_2 = pd.read_csv('/content/drive/MyDrive/emprestimos/emprestimos-20182.csv')
emprestimos2019_1 = pd.read_csv('/content/drive/MyDrive/emprestimos/emprestimos-20191.csv')
emprestimos2019_2 = pd.read_csv('/content/drive/MyDrive/emprestimos/emprestimos-20192.csv')
emprestimos2020_1 = pd.read_csv('/content/drive/MyDrive/emprestimos/emprestimos-20201.csv')
```

## **Concatenando DataFrames**

```python
BibliotecaDeEmprestimos = pd.concat([emprestimos2010_1, emprestimos2010_2, emprestimos2011_1, emprestimos2011_2,
                          emprestimos2012_1, emprestimos2012_2, emprestimos2013_1, emprestimos2013_2,
                          emprestimos2014_1, emprestimos2014_2, emprestimos2015_1, emprestimos2015_2,
                          emprestimos2016_1, emprestimos2016_2, emprestimos2017_1, emprestimos2017_2,
                          emprestimos2018_1, emprestimos2018_2, emprestimos2019_1, emprestimos2019_2,
                          emprestimos2020_1], ignore_index=True)
display(BibliotecaDeEmprestimos)
```

## **Filtrando as duplicatas**

```python
BibliotecaDeEmprestimos = BibliotecaDeEmprestimos.drop_duplicates()
BibliotecaDeEmprestimos.value_counts()
display(BibliotecaDeEmprestimos)
```

## **Importando Dados Exemplares**

```python
DadosExemplares = pd.read_parquet('/content/drive/MyDrive/emprestimos/dados_exemplares.parquet')
display(DadosExemplares)
```

##**Mesclando Dados**

```python
DadosCompletos = BibliotecaDeEmprestimos.merge(DadosExemplares)
display(DadosCompletos)
```

## **Manipulando os Dados corretamente**

Adicionando célula de CDU

```python
lista_CDU = []

for CDU in DadosCompletos['localizacao']:
 if (CDU < 100):
  lista_CDU.append('Generalidades')
 elif (CDU < 200):
  lista_CDU.append('Filosofia e psicologia')
 elif (CDU < 300):
  lista_CDU.append('Religião')
 elif (CDU < 400):
  lista_CDU.append('Ciências sociais')
 elif (CDU< 500):
  lista_CDU.append('Classe vaga')
 elif (CDU < 600):
  lista_CDU.append('Matemática e ciências naturais')
 elif (CDU < 700):
  lista_CDU.append('Ciências aplicadas')
 elif (CDU < 800):
  lista_CDU.append('Belas artes')
 elif (CDU < 900):
  lista_CDU.append('Linguagens')
 else:
  lista_CDU.append('Geografia e historia')
DadosCompletos['CDU'] = lista_CDU
display(DadosCompletos)
```

Modificando e Excluindo células desnecessárias

```python
DadosCompletos.drop(columns=['registro_sistema'], inplace=True)

DadosCompletos['matricula_ou_siape'] = DadosCompletos['matricula_ou_siape'].astype(str)
display(DadosCompletos)
```

Anlisando diferença entre empréstimos normais e exemplares

```python
# Merge the two DataFrames
merged_df = BibliotecaDeEmprestimos.merge(DadosExemplares, on='codigo_barras', how='left')

# Group by 'id_emprestimo' and count the number of unique 'id_exemplar'
exemplar_counts = merged_df.groupby('id_emprestimo')['id_exemplar'].nunique()

# Filter for loans with more than one exemplar ID
loans_with_multiple_exemplars = exemplar_counts[exemplar_counts > 1]

if not loans_with_multiple_exemplars.empty:
    print("There are loans with more than one exemplar ID:")
    display(loans_with_multiple_exemplars)
else:
    print("There are no loans with more than one exemplar ID.")
```

Livros exemplares emprestados por ano

```python
dataEmprestimos = pd.DataFrame(DadosCompletos['data_emprestimo'].value_counts()).reset_index()
dataEmprestimos.columns = ['Data', 'Quantidade']
dataEmprestimos['Data'] = pd.to_datetime(dataEmprestimos['Data'])
dataEmprestimos
```

```python
emprestimosPorAno = dataEmprestimos.groupby(dataEmprestimos['Data'].dt.year)['Quantidade'].sum().reset_index()
emprestimosPorAno.index.name = 'ano'
emprestimosPorAno
```
