## **Concatenando DataFrames**

```python
BibliotecaDeEmprestimos = pd.concat([emprestimos2010_1, emprestimos2010_2, emprestimos2011_1, emprestimos2011_2,
                          emprestimos2012_1, emprestimos2012_2, emprestimos2013_1, emprestimos2013_2,
                          emprestimos2014_1, emprestimos2014_2, emprestimos2015_1, emprestimos2015_2,
                          emprestimos2016_1, emprestimos2016_2, emprestimos2017_1, emprestimos2017_2,
                          emprestimos2018_1, emprestimos2018_2, emprestimos2019_1, emprestimos2019_2,
                          emprestimos2020_1], ignore_index=True)
BibliotecaDeEmprestimos
```

## **Filtrando as duplicatas**

```python
BibliotecaDeEmprestimos = BibliotecaDeEmprestimos.drop_duplicates()
BibliotecaDeEmprestimos.value_counts()
BibliotecaDeEmprestimos
```

## **Importando Dados Exemplares**

```python
DadosExemplares = pd.read_parquet('/content/drive/MyDrive/emprestimos/dados_exemplares.parquet')
DadosExemplares
```

##**Mesclando Dados**

```python
DadosCompletos = BibliotecaDeEmprestimos.merge(DadosExemplares)
DadosCompletos
```

## **Organizando os Dados corretamente**

Adicionando célula de CDU

```python
lista_CDU = []

for CDU in DadosCompletos['localizacao']:
 if (CDU < 100):
  lista_CDU.append('Generalidades')
 elif (CDU < 200):
  lista_CDU.append('Filosofia e psicologia')
 elif (CDU < 300):
  lista_CDU.append('Religião')
 elif (CDU < 400):
  lista_CDU.append('Ciências sociais')
 elif (CDU< 500):
  lista_CDU.append('Classe vaga')
 elif (CDU < 600):
  lista_CDU.append('Matemática e ciências naturais')
 elif (CDU < 700):
  lista_CDU.append('Ciências aplicadas')
 elif (CDU < 800):
  lista_CDU.append('Belas artes')
 elif (CDU < 900):
  lista_CDU.append('Linguagens')
 else:
  lista_CDU.append('Geografia e historia')
DadosCompletos['CDU'] = lista_CDU
DadosCompletos
```

Modificando e Excluindo células desnecessárias

```python
DadosCompletos.drop(columns=['registro_sistema'], inplace=True)

DadosCompletos['matricula_ou_siape'] = DadosCompletos['matricula_ou_siape'].astype(str)
DadosCompletos
```

Livros exemplares emprestados por ano

```python
dataEmprestimos = pd.DataFrame(DadosCompletos['data_emprestimo'].value_counts()).reset_index()
dataEmprestimos.columns = ['Data', 'Quantidade']
dataEmprestimos['Data'] = pd.to_datetime(dataEmprestimos['Data'])

# Organizando a quantidade dos emprestimos dos livros por ano
emprestimosPorAno = dataEmprestimos.groupby(dataEmprestimos['Data'].dt.year)['Quantidade'].sum().reset_index()
emprestimosPorAno
```

## **Formulando gráfico para análise dos dados anuais**

```python
plt.figure(figsize=(10, 6))
plt.plot(emprestimosPorAno['Data'], emprestimosPorAno['Quantidade'], marker='o')
plt.xlabel('Ano')
plt.ylabel('Quantidade')
plt.title('Quantidade de exemplares emprestados por ano')
plt.xticks(emprestimosPorAno['Data'], rotation=30)
plt.grid(True)
plt.show()
```

### **Relatório da análise anual**

Pode-se afirmar que a partir do ano de 2013 o número de empréstimos de livros feitos pela universidade decaiu, com ênfase no ano de 2019 que ocorreu uma queda bem grande desses exemplares.

Com isso, é necessária a observação cuidadosa por parte da biblioteca com relação ao sistema de empréstimos de livros após 2013, a fim de encontrar o motivo da diminuição desses empréstimos.


## **Verificando empréstimos feitos por mês**

```python
emprestimosPorMes = dataEmprestimos.groupby(dataEmprestimos['Data'].dt.month)['Quantidade'].sum().reset_index()

# trocando numeros por meses
meses = {1:'Jan', 2:'Fev', 3:'Mar', 4:'Abril', 5:'Maio', 6:'Jun', 7:'Jul', 8:'Ago', 9:'Set', 10:'Out', 11:'Nov', 12:'Dez'}
emprestimosPorMes['Data'] = emprestimosPorMes['Data'].map(meses)

emprestimosPorMes.columns = ['Mês', 'Quantidade']
emprestimosPorMes
```

**Formulando gráfico para análise dos dados mensais**

```python
plt.figure(figsize=(10, 6))
plt.plot(emprestimosPorMes['Mês'], emprestimosPorMes['Quantidade'], marker='o')
plt.ylabel('Quantidade')
plt.title('Quantidade de exemplares emprestados por mês (entre 2010 e 2020)')
plt.grid(True)
plt.show()
```

### **Relatório da análise dos dados mensais**

Percebe-se que o ápice dos empréstimos foram nos meses de **Março** e **Agosto**. Além disso, foram constatadas grandes quedas nos períodos entre **Junho** - **Julho** e **Dezembro** - **Janeiro**, devido as férias no período letivo.   

Até o momento os dados não demonstram anormalidades, contudo, existem estratégias que a diretoria da biblioteca pode adotar para aumentar o número de empréstimos durante os meses de baixa, como a organização de eventos e atividades extracurriculares, ou ações de marketing para estimular a leitura durante as férias.

## **Verificando empréstimos feitos por hora**

```python
emprestimosPorHora = dataEmprestimos.groupby(dataEmprestimos['Data'].dt.hour)['Quantidade'].sum().reset_index()
emprestimosPorHora.columns = ['Hora', 'Quantidade']
emprestimosPorHora
```

**Formulando gráfico para análise dos dados por hora**

```python
plt.figure(figsize=(10, 6))
bar = sns.barplot(data=emprestimosPorHora,y='Quantidade',x='Hora',
                 palette='Blues',hue='Quantidade',dodge=False)

plt.legend([],[], frameon=False)

bar.set(xlabel='Horário',ylabel='Quantidade')
bar.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',').replace(',','.')))
bar.set_title("Quantidade de exemplares emprestados por faixa horária"+"\n",size=15,loc='left',weight='bold')
bar.text(s='Período de 2010 a 2020',x=-0.5,y=225000,fontsize=12, ha='left',color='gray')
bar=bar
```

### **Relatório da análise dos dados por hora**

Através do gráfico é possível identificar que os períodos entre às 10h e 16h são os de maior pico no atendimento de empréstimos.   

Além disso, os períodos do começo da manhã (entre 7h e 9) e da noite (entre 19h e 22) podem ser os melhores horários para que se realizem as demais atividades além do atendimento.

###**Calculando a frequência das variáveis**

```python
def Valorfrequencia(variavel):

  frequencia = pd.DataFrame(DadosCompletos[variavel].value_counts())
  frequencia.columns = ['Quantidade']
  frequencia['Porcentagem'] = round(frequencia.Quantidade / frequencia.Quantidade.sum() * 100, 1)

  return frequencia
```

### Analisando as frequências de cada variável

#### **Frequência dos tipos de vinculo por usuario**

```python
Valorfrequencia('tipo_vinculo_usuario')
```

```python
frequencia_tipo_vinculo = Valorfrequencia('tipo_vinculo_usuario')

plt.figure(figsize=(12, 7))
bar = sns.barplot(data=frequencia_tipo_vinculo, y='Porcentagem', x=frequencia_tipo_vinculo.index, palette='viridis', hue='Porcentagem', dodge=False)

plt.legend([], [], frameon=False)
plt.xlabel('Tipo de Vínculo do Usuário')
plt.ylabel('Porcentagem')
plt.title('Frequência de Empréstimos por Tipo de Vínculo de Usuário')
plt.xticks(rotation=45, ha='right')
bar.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',').replace(',','.')))
plt.tight_layout()
plt.show()
```

#### **Análise da frequência dos tipos de usuários**   

Percebe-se que o grande volume de empréstimos se deve aos alunos de graduação, responsáveis por, praticamente, 80% dos empréstimos totais. Em seguida, os alunos de pós-graduação representam o segundo maior público com cerca de 15% dos empréstimos totais.

Possilvemente a maior diferença entre os percentuais dos dois públicos anteriores é devido a dois pontos:

- Alunos de pós graduação, priorizam estudar por artigos científicos que são acessados através de bases de dados bibliográficas.

- A grande diferença no número de alunos de graduação em comparação com os da pós na universidade.
